En este reto deberás mezclar los canales rojo, verde y azul: Muestra la mezcla de los canales rojo-verde, rojo-azul, y verde-azul. ¿Que observas de interesante?

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

imageOpenCV = cv2.imread("mexicocity.jpg", cv2.IMREAD_COLOR)
imagen = cv2.cvtColor(imageOpenCV, cv2.COLOR_BGR2RGB)
```

Separamos los tres canales de color y generamos un canal "apagado" con ceros. 

```python
canal_rojo = imagen[:,:,0]
canal_verde = imagen[:,:,1]
canal_azul = imagen[:,:,2]

#Volveremos los canales que ignoramos a 0.
canalApagado = np.zeros_like(canal_rojo)
```

### Canal Rojo y Verde

```python
img_canal_rojoverde = np.dstack((canal_rojo, canal_verde, canalApagado))
print(img_canal_rojoverde.shape)
plt.figure(figsize=(40,40))
plt.imshow(img_canal_rojoverde)
```

### Canal Rojo y Azul

```python
img_canal_rojoazul = np.dstack((canal_rojo, canalApagado, canal_azul))
print(img_canal_rojoazul.shape)
plt.figure(figsize=(40,40))
plt.imshow(img_canal_rojoazul)
```

### Canal Verde y Azul

```python
img_canal_verdeazul = np.dstack((canalApagado, canal_verde, canal_azul))
print(img_canal_verdeazul.shape)
plt.figure(figsize=(40,40))
plt.imshow(img_canal_verdeazul)
```

Como puedes notar, los colores son Cyan, Magenta y Amarillo, estos tres colores, combinados con el negro, son los que necesitas para imprimir una imagen en colores.
